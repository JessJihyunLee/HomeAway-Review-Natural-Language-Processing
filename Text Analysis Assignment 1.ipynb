B1. In this section, you will create classification models to predict high (75th percentile and above) or low(below 75th percentile) salary from the text contained in the job descriptions. Ignore the job descriptions, and train a model to predict high/low salary from all the numeric columns, e.g., part time/full time, contract vs. others, type of job (a lot of dummy variables), location (instead of using a huge number of dummy variables, you can use a list of cities in England with highest cost of living, and create a 0/1 variable which is 1 if the job is in one of those cities, else 0). Use the NaÃ¯ve Bayes classifier. What is the accuracy of your model?    

```python
import copy
data4 = copy.deepcopy(data1)
```

```python
data5 = data4.dropna()

high = data4['SalaryNormalized'].quantile(q=0.75)

data5['SalaryNormalized'][data4['SalaryNormalized']>=high] = 1
data5['SalaryNormalized'][data4['SalaryNormalized']<high] = 0

data5['ContractTime'][data4['ContractTime']=='contract'] = 'contract'
data5['ContractTime'][data4['ContractTime']!='contract'] = 'others'
```

Divided Location based on House Price from 'Annual Affordable Cities Review'
https://www.independent.co.uk/life-style/uk-least-affordable-cities-house-prices-london-oxford-a7606326.html

```python
def lowerize(x):
    return x.lower()

data5['LocationNormalized'] = data5['LocationNormalized'].apply(lowerize)
```

```python
highlivingcost = ['oxford','uk','the city','greater london','central london','london','south east london','winchester', 'cambridge', 'chichester', 'brighon and hove', 'bath', 'southhampton','salisbury','canterbury','st albans', 'bristol','lichfield','truro','norwich','chelmsford','exeter','york','leicester','gloucester']
mask2 = data5['LocationNormalized'].isin(highlivingcost)

data5['LocationNormalized'][mask2] = 1
data5['LocationNormalized'][mask2 == False] = 0
data5.drop(columns=['Id','FullDescription','LocationRaw','SalaryRaw','Title','Company','SourceName'])
```

Create the Design Matrices

```python
from patsy import dmatrices

categorical_columns = ['ContractType','ContractTime','Category']
data5_dummies = pd.get_dummies(data5[categorical_columns],
                            prefix=categorical_columns,
                            columns=categorical_columns)
dummy_column_names = data5_dummies.columns.values

data6 = pd.concat([data5, data5_dummies], axis=1)
```

```python
formula = 'SalaryNormalized ~ 0 + LocationNormalized + {}'.format(' + '.join(['Q("{}")'.format(x) for x in dummy_column_names]))

Y, X = dmatrices(formula, data6, return_type = 'dataframe')
y = Y['SalaryNormalized'].values
```

Split into Train and Test Data

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=1)
```

Set up the Classifier, Fit the Model, and Test on Training, Test Data

```python
from sklearn import naive_bayes
from sklearn import metrics

model = naive_bayes.MultinomialNB()
model.fit(X_train, y_train)

prediction_train = model.predict(X_train)
print(metrics.confusion_matrix(y_train, prediction_train))
print(metrics.accuracy_score(y_train, prediction_train))
```

```python
prediction_test = model.predict(X_test)
print(metrics.confusion_matrix(y_test, prediction_test))
print(metrics.accuracy_score(y_test, prediction_test))
```

Class Priors(or Baseline)

```python
import math

print('Prior probability for positive class')
print(math.exp(model.class_log_prior_[1]))
print('Prior probability for negative class')
print(math.exp(model.class_log_prior_[0]))
```
